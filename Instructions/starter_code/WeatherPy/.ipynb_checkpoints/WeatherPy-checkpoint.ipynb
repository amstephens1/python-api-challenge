```python
import json
import requests
from pprint import pprint
import numpy as np
import pandas as pd
from api_keys import weather_api_key
from citipy import citipy
from matplotlib import pyplot as plt
from scipy.stats import linregress
```

## Generate Cities List

```python
rand_lat = np.random.uniform(-90, 90, 1750)
rand_lng = np.random.uniform(-180, 80, 1750)

cities_df = pd.DataFrame({"Latitude": rand_lat, "Longitude": rand_lng})
cities_df.head()
```

```python
# Test Code
# city = citipy.nearest_city(3.629137, -62.674587)
# city.city_name 
# city.country_code
```

```python
city = []
country_code = []

for x in range(0, len(cities_df["Latitude"])):
    selected_city = citipy.nearest_city(cities_df.loc[x]["Latitude"], cities_df.iloc[x]["Longitude"]).city_name
    city.append(selected_city)
    
    selected_country = citipy.nearest_city(cities_df.loc[x]["Latitude"], cities_df.iloc[x]["Longitude"]).country_code
    country_code.append(selected_country)

cities_df["City"] = city
cities_df["Country Code"] = country_code
cities_df
  
```

```python
# Confirm there are more than 500 unique cities
cities_dup_df = cities_df.drop_duplicates("City", keep="first")
cities_sample_df = cities_dup_df.reset_index(drop=True)
cities_sample_df
```

### Perform API Calls
* Perform a weather check on each city using a series of successive API calls.
* Include a print log of each city as it'sbeing processed (with the city number and city name).


```python
#Test Code
# url = "http://api.openweathermap.org/data/2.5/weather?"
# city = "London"
# query_url = url + "appid=" + weather_api_key + "&q=" + city

# weather_response = requests.get(query_url)
# weather_json = weather_response.json()

# # Get the temperature from the response
# print(f"The weather API responded with: {weather_json}.")
```

```python
url = "http://api.openweathermap.org/data/2.5/weather?"

lat = []
lng = []
max_temp = []
humidity = []
cloudiness = []
wind_speed = []
date = []
count = 1
units = "imperial"

for index, row in cities_sample_df.iterrows():
    
    city = row["City"]
    country_code = row["Country Code"]
    count = count + 1
    query_url = f"{url}q={city},{country_code}&units={units}&appid={weather_api_key}"
    weather_response = requests.get(query_url).json()
    
    try:
        lat.append(weather_response['coord']['lat'])
        lng.append(weather_response['coord']['lon'])
        max_temp.append(weather_response['main']['temp_max'])
        humidity.append(weather_response['main']['humidity'])
        cloudiness.append(weather_response['clouds']['all'])
        wind_speed.append(weather_response['wind']['speed'])
        date.append(weather_response['dt'])
        print(f"{city} has been added to the log. This is city number {count}")
        count = count + 1
    except:
        lat.append(np.nan)
        lng.append(np.nan)
        max_temp.append(np.nan)
        humidity.append(np.nan)
        wind_speed.append(np.nan)
        cloudiness.append(np.nan)
        date.append(np.nan)
        print(f"Missing field/result for {city}, skipping.")
```

```python
cities_sample_df["Latitude"] = lat
cities_sample_df["Longitude"] = lng
cities_sample_df["Max Temp"] = max_temp
cities_sample_df["Humidity"] = humidity
cities_sample_df["Cloudiness"] = cloudiness
cities_sample_df["Wind Speed"] = wind_speed
cities_sample_df["Date"] = date
cities_sample_df
```

```python
weather_check_df = cities_sample_df.dropna().reset_index(drop=True)
weather_check_df
```

## Linear Regression

```python
#Split data into Northern and Southern Hemispheres
northern_hemisphere = weather_check_df[weather_check_df["Latitude"]>0]
southern_hemisphere = weather_check_df[weather_check_df["Latitude"]<0]
weather_check_df
```

####  Northern Hemisphere - Max Temp vs. Latitude Linear Regression

```python
x_values = northern_hemisphere['Latitude']
y_values = northern_hemisphere['Max Temp']
(slope, intercept, rvalue, pvalue, stderr) = linregress(x_values, y_values)
regress_values = x_values * slope + intercept
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
plt.scatter(x_values,y_values)
plt.plot(x_values,regress_values,"r-")
plt.annotate(line_eq,(6,10),fontsize=15,color="red")
plt.title("Northern Hemisphere Latitude vs. Temperature Linear Regression")
plt.xlabel('Latitude')
plt.ylabel('Max Temperature (F)')
print(f"The r-squared is: {rvalue**2}")
plt.show()
```

####  Southern Hemisphere - Max Temp vs. Latitude Linear Regression

```python
x_values = southern_hemisphere['Latitude']
y_values = southern_hemisphere['Max Temp']
(slope, intercept, rvalue, pvalue, stderr) = linregress(x_values, y_values)
regress_values = x_values * slope + intercept
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
plt.scatter(x_values,y_values)
plt.plot(x_values,regress_values,"r-")
plt.annotate(line_eq,(-50,80),fontsize=15,color="red")
plt.title("Southern Hemisphere Latitude vs. Temperature Linear Regression")
plt.xlabel('Latitude')
plt.ylabel('Max Temperature (F)')
print(f"The r-squared is: {rvalue**2}")
plt.show()
```

```python
# There is a correlation between the increas in temperature as you move closer to the equator.
# The northern plot is a stronger correlation, and I would guess that is due to the much higher sample size. 
```
